# Xenograft In Vivo Efficacy Analysis

Analyses included:
1. **Primary**: Linear Mixed-Effects model (LME) on log tumor volumes (Group × Day) with spaghetti plot and mean±95% CI.
2. **Secondary**: Integrated tumor burden (**AUC**) per mouse with group comparison (Kruskal–Wallis + pairwise Wilcoxon) and swarm/violin plot.
3. **Secondary**: **Kaplan–Meier** time-to-endpoint with log-rank tests and **Cox** hazard ratios.
4. **Snapshot**: **TGI% at Day 28** (configurable) with lollipop plot per mouse.

**Input file**: `xenograft_volume.csv`. Place it in the **current working directory**.

```r
# Packages
req_pkgs <- c(
  'tidyverse','readr','janitor','stringr','lubridate',
  'lmerTest','emmeans','ggeffects',
  'survival','survminer',
  'pracma','DescTools'
)
for (p in req_pkgs) {
  if (!requireNamespace(p, quietly = TRUE)) install.packages(p, repos = 'https://cloud.r-project.org')
}
invisible(lapply(req_pkgs, library, character.only = TRUE))

theme_set(theme_minimal(base_size = 12))
```

```r
# Config
# Humane endpoint (mm^3) for KM analysis
ENDPOINT_MM3 <- 150
# Snapshot day for TGI
SNAPSHOT_DAY <- 45

# Output directory for figures
outdir <- 'figures'
if (!dir.exists(outdir)) dir.create(outdir)
```

```r
library(tidyverse)
library(janitor)
library(stringr)
library(readr)

# Load & Tidy Data (robust to x-prefixed day columns) ----
raw <- read_csv("xenograft_volume.csv", show_col_types = FALSE) |> clean_names()

# sanity check
stopifnot("sample" %in% names(raw))

dat_long <- raw |>
  mutate(sample = str_squish(sample)) |>
  pivot_longer(
    cols = -all_of("sample"),
    names_to = "day_chr",
    values_to = "volume"
  ) |>
  mutate(
    # parse numeric day even if name is "x0", "day_28", "D28", etc.
    day    = readr::parse_number(day_chr),
    volume = as.numeric(volume)
  ) |>
  arrange(sample, day) |>
  filter(!is.na(day))

# derive Group from sample names (your labels)
dat_long <- dat_long |>
  mutate(
    group = case_when(
      str_detect(sample, "^Anti-HER2/neu Construct #\\s*1") ~ "anti-HER2 Construct #1",
      str_detect(sample, "^Anti-HER2/neu Construct #\\s*2") ~ "anti-HER2 Construct #2",
      str_detect(sample, "^Anti-HER2/neu Construct #\\s*3") ~ "anti-HER2 Construct #3",
      str_detect(sample, "^Non-TD")                         ~ "Non-TD",
      str_detect(sample, "^Intact")                         ~ "Intact",
      TRUE                                                  ~ "Other"
    ),
    mouse_id = factor(sample),
    log_vol  = log1p(volume)
  )

# Diagnostics so you can see what’s happening
message("Columns in CSV (after clean_names):")
print(names(raw))
message("Row counts by group:")
print(dat_long |> count(group, name = "n_rows"))
message("Mice per group:")
print(dat_long |> summarise(n_mice = n_distinct(mouse_id), .by = group))
message("Day range per group:")
print(dat_long |> summarise(min_day = min(day), max_day = max(day), .by = group))
```

## Primary: LME on log volumes (Group × Day)

```r
# LME model (Group × Day)
library(lmerTest)
library(emmeans)
library(dplyr)
library(ggplot2)

# 1) Keep groups of interest (must match your derived labels)
groups_keep <- c("anti-HER2 Construct #1",
                 "anti-HER2 Construct #2",
                 "anti-HER2 Construct #3",
                 "Non-TD","Intact")

d_lme <- dat_long %>%
  filter(group %in% groups_keep) %>%
  tidyr::drop_na(day, volume, log_vol) %>%
  arrange(group, mouse_id, day) %>%
  droplevels()

# Sanity checks
print(table(d_lme$group))
stopifnot(nrow(d_lme) > 0, dplyr::n_distinct(d_lme$mouse_id) >= 2)

# 2) Center day for numerical stability
d_lme <- d_lme %>% mutate(day_c = day - mean(day, na.rm = TRUE))

# 3) Fit LME: log(volume+1) ~ Group * Day + random intercept & slope per mouse
m_lme <- lmer(log_vol ~ group * day_c + (day_c | mouse_id), data = d_lme)
summary(m_lme)
anova_lme <- anova(m_lme, type = 3)   # Type III for interaction
anova_lme

# 4) Predicted mean trajectories with 95% CI via emmeans (on log scale), then back-transform
day_seq <- seq(min(d_lme$day_c, na.rm = TRUE), max(d_lme$day_c, na.rm = TRUE), by = 1)

emm <- emmeans(m_lme, ~ group | day_c, at = list(day_c = day_seq))
emm_df <- as.data.frame(emm) %>%
  mutate(
    day    = day_c + mean(d_lme$day, na.rm = TRUE),
    fit    = exp(emmean) - 1,
    lwr    = exp(lower.CL) - 1,
    upr    = exp(upper.CL) - 1
  )

# 5) Plot: spaghetti (per mouse) + mean±95% CI ribbons/lines (log y-axis)
p_lme <- ggplot() +
  geom_line(data = d_lme,
            aes(x = day, y = volume, group = mouse_id, color = group),
            alpha = 0.25) +
  geom_ribbon(data = emm_df,
              aes(x = day, ymin = lwr, ymax = upr, fill = group),
              alpha = 0.20, colour = NA) +
  geom_line(data = emm_df,
            aes(x = day, y = fit, color = group),
            linewidth = 1.1) +
  scale_y_continuous(trans = "log10", labels = scales::comma) +
  labs(x = "Day",
       y = "Tumor volume (mm^3, log scale)",
       color = "Group", fill = "Group",
       title = "Tumor growth kinetics: LME (log volumes) with mean±95% CI and spaghetti") +
  theme_minimal(base_size = 12) +
  theme(legend.position = "right")

print(p_lme)
ggsave(file.path(outdir, "01_LME_log_volume_spaghetti_meanCI.png"),
       p_lme, width = 9, height = 6, dpi = 300)
```

## Secondary: Integrated Tumor Burden (AUC) per mouse

```r
#4) AUC computation & group comparison
# Use log1p(volume) for iAUC as discussed
auc_df <- d_lme %>% 
  arrange(mouse_id, day) %>%
  group_by(mouse_id, group) %>%
  summarise(iAUC = DescTools::AUC(x = day, y = log_vol, method = 'trapezoid'), .groups='drop')

# Non-parametric comparison across groups
kw <- kruskal.test(iAUC ~ group, data = auc_df)
kw

pairwise <- pairwise.wilcox.test(auc_df$iAUC, auc_df$group, p.adjust.method = 'BH')
pairwise

p_auc <- ggplot(auc_df, aes(group, iAUC)) +
  geom_violin(trim = FALSE, alpha = 0.2) +
  geom_boxplot(width = 0.15, outlier.shape = NA) +
  ggbeeswarm::geom_quasirandom(width = 0.2, alpha = 0.7) +
  labs(x = 'Group', y = 'Integrated tumor burden (AUC of log1p volume)',
       title = sprintf('AUC comparison across groups (Kruskal–Wallis p = %.3g)', kw$p.value)) +
  theme(axis.text.x = element_text(angle = 30, hjust = 1))
p_auc
ggsave(file.path(outdir, '02_AUC_violin_swarm.png'), p_auc, width = 8, height = 6, dpi = 300)
```

## Secondary: Kaplan–Meier survival, log-rank, and Cox HR

```r
# Time-to-endpoint derivation
# Define time-to-event: earliest day with volume >= ENDPOINT_MM3; otherwise censor at last day observed
km_df <- d_lme %>% group_by(mouse_id, group) %>% summarise(
  time = {
    above <- day[which(volume >= ENDPOINT_MM3)]
    if (length(above) > 0) min(above) else max(day, na.rm = TRUE)
  },
  status = as.integer(any(volume >= ENDPOINT_MM3)),
  .groups='drop'
)

fit <- survfit(Surv(time, status) ~ group, data = km_df)
p_km <- ggsurvplot(
  fit, data = km_df, pval = TRUE, conf.int = TRUE, risk.table = TRUE,
  legend.title = 'Group', legend = 'right',
  xlab = 'Days', ylab = 'Survival (below endpoint)',
  ggtheme = theme_minimal(base_size = 12)
)
p_km
ggsave(file.path(outdir, '03_KM_logrank.png'), p_km$plot, width = 8, height = 6, dpi = 300)
ggsave(file.path(outdir, '03_KM_logrank_risktable.png'), p_km$table, width = 8, height = 3, dpi = 300)

# Cox model for hazard ratios (vs reference group: Intact if present, else first level)
km_df$group <- relevel(factor(km_df$group), ref = ifelse('Intact' %in% km_df$group, 'Intact', levels(factor(km_df$group))[1]))
cox <- coxph(Surv(time, status) ~ group, data = km_df)
summary(cox)
```

## Snapshot: TGI% at a chosen day

```r
# TGI% calculation and lollipop plot
d_snap <- d_lme %>% filter(day %in% c(0, SNAPSHOT_DAY)) %>%
  select(mouse_id, group, day, volume)

# Restrict to mice with both baseline and snapshot values
d_wide <- d_snap %>% pivot_wider(names_from = day, values_from = volume, names_prefix = 'day_') %>%
  drop_na(day_0, !!sym(paste0('day_', SNAPSHOT_DAY))) %>%
  mutate(delta = !!sym(paste0('day_', SNAPSHOT_DAY)) - day_0)

# Control = Intact mean delta; fallback to Non-TD if Intact absent
ctrl_grp <- if ('Intact' %in% d_wide$group) 'Intact' else if ('Non-TD' %in% d_wide$group) 'Non-TD' else levels(factor(d_wide$group))[1]
delta_ctrl <- d_wide %>% filter(group == ctrl_grp) %>% summarise(mean_delta = mean(delta, na.rm = TRUE)) %>% pull(mean_delta)

tgi_df <- d_wide %>% mutate(
  TGI = 100 * (1 - (delta / delta_ctrl))
) %>% arrange(group, desc(TGI))

p_tgi <- ggplot(tgi_df, aes(x = reorder(mouse_id, TGI), y = TGI, color = group)) +
  geom_segment(aes(xend = reorder(mouse_id, TGI), y = 0, yend = TGI), linewidth = 0.8) +
  geom_point(size = 2) +
  coord_flip() +
  labs(x = 'Mouse', y = sprintf('TGI at day %d (%%)', SNAPSHOT_DAY),
       title = sprintf('TGI per mouse at day %d (control = %s, meanΔ=%.1f)', SNAPSHOT_DAY, ctrl_grp, delta_ctrl))
p_tgi
ggsave(file.path(outdir, '04_TGI_lollipop.png'), p_tgi, width = 8, height = 6, dpi = 300)
```
