# Xenograft in vivo treatment efficacy

**Outputs**
- Linear-scale growth curves (combined + per-biogroup panels + per-group files)
- AUC over all days; KW + pairwise Wilcoxon
- TGI% at Day 45 and AUC-based TGI (uses full window)

**Input**: `xenograft_volume.csv`

```r
# Packages
req <- c('tidyverse','janitor','stringr','readr','scales',
         'DescTools','lmerTest','emmeans','ggbeeswarm','patchwork')
for (p in req) if (!requireNamespace(p, quietly = TRUE)) install.packages(p, repos='https://cloud.r-project.org')
invisible(lapply(req, library, character.only = TRUE))
theme_set(theme_minimal(base_size = 12))
```

```r
# 1) Config & helpers
SNAPSHOT_DAY <- 45
CTRL_FALLBACK <- 'Non-TD'
outdir <- 'figures'
if (!dir.exists(outdir)) dir.create(outdir)

save_jpeg <- function(plot, filename, width=9, height=6, dpi=600){
  ggsave(file.path(outdir, paste0(filename, '.jpeg')), plot,
         device='jpeg', width=width, height=height, dpi=dpi)
}
```

```r
# 2) Load & tidy data
raw <- readr::read_csv('xenograft_volume.csv', show_col_types = FALSE) |>
  janitor::clean_names()
stopifnot('sample' %in% names(raw))

dat_long <- raw |>
  mutate(sample = stringr::str_squish(sample)) |>
  pivot_longer(cols = -all_of('sample'), names_to = 'day_chr', values_to = 'volume') |>
  mutate(
    day    = readr::parse_number(day_chr),
    volume = as.numeric(volume)
  ) |>
  arrange(sample, day) |>
  filter(!is.na(day)) |>
  mutate(
    group = case_when(
      str_detect(sample, '^Anti-HER2/neu Construct #1') ~ 'anti-HER2 Construct #1',
      str_detect(sample, '^Anti-HER2/neu Construct #2') ~ 'anti-HER2 Construct #2',
      str_detect(sample, '^Anti-HER2/neu Construct #3') ~ 'anti-HER2 Construct #3',
      str_detect(sample, '^Non-TD') ~ 'Non-TD',
      str_detect(sample, '^Intact') ~ 'Intact',
      TRUE ~ 'Other'
    ),
    mouse_id = factor(sample),
    log_vol  = log1p(volume)
  )

groups_keep <- c('anti-HER2 Construct #1','anti-HER2 Construct #2','anti-HER2 Construct #3','Non-TD','Intact')
dat_long <- dat_long |> filter(group %in% groups_keep) |> droplevels()

dat_long |>
  summarise(n_mice = n_distinct(mouse_id), .by = group) |>
  print()
dat_long |>
  summarise(min_day=min(day), max_day=max(day), .by=group) |>
  print()
```

## 3) Linear-scale growth curves (combined + panels + per-group)

```r
# Growth curves (linear), 100 mm³ y-ticks, exact day x-ticks, JPEG 600 dpi -
library(lmerTest)
library(ggplot2)
library(dplyr)
library(scales)

# settings 
SHOW_SPAGHETTI <- FALSE              # set TRUE if you want spaghetti
day_breaks <- sort(unique(dat_long$day))  # ticks exactly at measurement days (0,3,6,9,...)

# prepare data for LME & summaries 
d_lme <- dat_long |>
  tidyr::drop_na(day, volume) |>
  mutate(day_c = day - mean(day, na.rm = TRUE)) |>
  arrange(group, mouse_id, day) |>
  droplevels()

stopifnot(nrow(d_lme) > 0, dplyr::n_distinct(d_lme$mouse_id) >= 2)

# Optional: LME for inference only (plots stay linear)
m_lme <- tryCatch(
  lmer(log1p(volume) ~ group * day_c + (day_c | mouse_id), data = d_lme),
  error = function(e) { message("LME failed: ", e$message); NULL }
)
if (!is.null(m_lme)) {
  print(anova(m_lme, type = 3))
}

# Per-day summaries for mean±SEM
sum_lin <- dat_long |>
  summarise(
    n    = dplyr::n(),
    mean = mean(volume, na.rm = TRUE),
    sd   = sd(volume, na.rm = TRUE),
    sem  = sd / sqrt(n),
    .by  = c(group, day)
  )

# Helper to make 100-step y-breaks up to the top of error bars
y_max_all <- max(sum_lin$mean + sum_lin$sem, na.rm = TRUE)
y_breaks_all <- seq(0, ceiling(y_max_all/100)*100, by = 100)

# Combined plot (mean±SEM; spaghetti optional) 
p_growth_combined <- ggplot()

if (SHOW_SPAGHETTI) {
  p_growth_combined <- p_growth_combined +
    geom_line(data = dat_long,
              aes(day, volume, group = mouse_id, color = group),
              alpha = 0.25, linewidth = 0.5)
}

p_growth_combined <- p_growth_combined +
  geom_line(data = sum_lin, aes(day, mean, color = group), linewidth = 1.1) +
  geom_errorbar(data = sum_lin,
                aes(day, ymin = pmax(mean - sem, 0), ymax = mean + sem, color = group),
                width = 0.8, linewidth = 0.4) +
  geom_point(data = sum_lin, aes(day, mean, color = group), size = 2) +
  scale_x_continuous(breaks = day_breaks) +
  scale_y_continuous(breaks = y_breaks_all, limits = c(0, max(y_breaks_all)),
                     labels = comma) +
  labs(x = "Day", y = "Tumor volume (mm³)", color = "Group",
       title = paste0("Tumor growth (linear): mean ± SEM",
                      if (SHOW_SPAGHETTI) " + spaghetti" else "")) +
  theme_minimal(base_size = 12) +
  theme(legend.position = "right",
        panel.grid.minor.x = element_blank())

print(p_growth_combined)
ggsave(file.path(outdir, "01_growth_linear_combined.jpeg"),
       p_growth_combined, device = "jpeg", width = 12, height = 7, dpi = 600)

# Clean per-biogroup panels (mean±SEM only) 

p_growth_panels <- ggplot(sum_lin, aes(day, mean)) +
  geom_line(color = "black", linewidth = 1.1) +
  geom_errorbar(aes(ymin = pmax(mean - sem, 0), ymax = mean + sem), width = 0.8) +
  geom_point(size = 2) +
  facet_wrap(~ group, scales = "free_y") +
  scale_x_continuous(breaks = day_breaks) +
  scale_y_continuous(breaks = scales::breaks_width(100), labels = comma) +
  labs(x = "Day", y = "Tumor volume (mm³)",
       title = "Tumor growth per group (mean ± SEM)") +
  theme_minimal(base_size = 12) +
  theme(panel.grid.minor.x = element_blank())

print(p_growth_panels)
ggsave(file.path(outdir, "02_growth_linear_panels.jpeg"),
       p_growth_panels, device = "jpeg", width = 14, height = 7, dpi = 600)
```

## 4) AUC over full window (0–max day)

```r
auc_df <- dat_long |>
  arrange(mouse_id, day) |>
  group_by(mouse_id, group) |>
  summarise(AUC = DescTools::AUC(x = day, y = volume, method='trapezoid'), .groups='drop')

kw <- kruskal.test(AUC ~ group, data = auc_df)
pair <- pairwise.wilcox.test(auc_df$AUC, auc_df$group, p.adjust.method='BH')
print(kw); print(pair)
```

## 5) TGI at Day 45 + AUC-based TGI

```r
# 1) Compute TGI per mouse at the chosen snapshot
snap_wide <- dat_long |>
  filter(day %in% c(0, SNAPSHOT_DAY)) |>
  select(mouse_id, group, day, volume) |>
  tidyr::pivot_wider(names_from = day, values_from = volume, names_prefix = "day_") |>
  tidyr::drop_na(day_0, !!sym(paste0("day_", SNAPSHOT_DAY))) |>
  mutate(delta = !!sym(paste0("day_", SNAPSHOT_DAY)) - day_0)

# control: prefer Intact, else fallback (e.g., Non-TD)
ctrl <- if ("Intact" %in% snap_wide$group) "Intact" else CTRL_FALLBACK
delta_ctrl <- snap_wide |>
  filter(group == ctrl) |>
  summarise(mean_delta = mean(delta, na.rm = TRUE), .groups = "drop") |>
  pull(mean_delta)

tgi_df <- snap_wide |>
  mutate(TGI = 100 * (1 - (delta / delta_ctrl))) |>
  arrange(group, desc(TGI))

# 2) Lollipop per mouse (like before), combined
p_tgi_snap <- ggplot(tgi_df, aes(x = reorder(mouse_id, TGI), y = TGI, color = group)) +
  geom_hline(yintercept = 0, linetype = "dashed") +
  geom_segment(aes(xend = reorder(mouse_id, TGI), y = 0, yend = TGI), linewidth = 0.8) +
  geom_point(size = 2) +
  coord_flip() +
  labs(
    x = "Mouse",
    y = sprintf("TGI at day %s (%%)", SNAPSHOT_DAY),
    title = sprintf("Snapshot TGI per mouse (control = %s; meanΔ(control) = %.1f mm³)", ctrl, delta_ctrl)
  ) +
  theme_minimal(base_size = 12)

p_tgi_snap
ggsave(file.path(outdir, "tgi_snapshot_lollipop.jpeg"),
       p_tgi_snap, device = "jpeg", width = 8, height = 10, dpi = 600)

# Standard palette to match your attached growth plot (ggplot2 default discrete)
pal_std <- c(
  "Intact"                     = "#F8766D",  # red
  "Non-TD"                     = "#B79F00",  # olive
  "anti-HER2 Construct #1"     = "#00BA38",  # green
  "anti-HER2 Construct #2"     = "#00BFC4",  # cyan
  "anti-HER2 Construct #3"     = "#C77CFF"   # purple
)

p_tgi_box <- ggplot(tgi_df, aes(group, TGI, fill = group)) +
  geom_boxplot(outlier.shape = NA, alpha = 0.25, width = 0.6, color = "grey40") +
  ggbeeswarm::geom_quasirandom(
    aes(fill = group),
    width = 0.2, alpha = 0.9,
    shape = 21, color = "black", size = 2.8   # filled points with black edge
  ) +
  geom_hline(yintercept = 0, linetype = "dashed") +
  scale_fill_manual(values = pal_std, guide = "none") +   # <- palette here
  scale_color_manual(values = pal_std, guide = "none") +  # (safety if any layer uses color)
  labs(
    x = "Group",
    y = sprintf("TGI at day %s (%%)", SNAPSHOT_DAY),
    title = sprintf("Snapshot TGI by group (control = %s)", ctrl)
  ) +
  theme_minimal(base_size = 12) +
  theme(axis.text.x = element_text(angle = 20, hjust = 1))

p_tgi_box
ggsave(file.path(outdir, "tgi_snapshot_box_swarm.jpeg"),
       p_tgi_box, device = "jpeg", width = 8.5, height = 5, dpi = 600)


snap_wide <- dat_long |>
  filter(day %in% c(0, SNAPSHOT_DAY)) |>
  select(mouse_id, group, day, volume) |>
  tidyr::pivot_wider(names_from = day, values_from = volume, names_prefix = 'day_') |>
  drop_na(day_0, !!sym(paste0('day_', SNAPSHOT_DAY))) |>
  mutate(delta = !!sym(paste0('day_', SNAPSHOT_DAY)) - day_0)

ctrl <- if ('Intact' %in% snap_wide$group) 'Intact' else CTRL_FALLBACK
delta_ctrl <- snap_wide |> filter(group == ctrl) |> summarise(mean_delta = mean(delta, na.rm=TRUE)) |> pull(mean_delta)

tgi_df <- snap_wide |> mutate(TGI = 100 * (1 - (delta / delta_ctrl))) |> arrange(group, desc(TGI))

auc_df <- dat_long |>
  arrange(mouse_id, day) |>
  group_by(mouse_id, group) |>
  summarise(AUC = DescTools::AUC(day, volume, method='trapezoid'), .groups='drop')
auc_ctrl_mean <- auc_df |> filter(group == ctrl) |> summarise(mean_auc = mean(AUC, na.rm=TRUE)) |> pull(mean_auc)
tgi_auc <- auc_df |> mutate(TGI_AUC = 100 * (1 - (AUC / auc_ctrl_mean)))

p_tgi_auc <- ggplot(tgi_auc, aes(x = reorder(mouse_id, TGI_AUC), y = TGI_AUC, color = group)) +
  geom_hline(yintercept = 0, linetype = 'dashed') +
  geom_segment(aes(xend = reorder(mouse_id, TGI_AUC), y = 0, yend = TGI_AUC), linewidth = 0.8) +
  geom_point(size = 2) +
  coord_flip() +
  labs(x='Mouse', y='TGI based on AUC (%)', title=sprintf('AUC-based TGI (control = %s; mean AUC = %.1f)', ctrl, auc_ctrl_mean))
p_tgi_auc
ggsave(file.path(outdir, 'tgi_auc_lollipop.jpeg'), p_tgi_auc, device='jpeg', width=8, height=10, dpi=600)
```
